```python
def faixasph(df):
    x=df['V9532']
    if x<1000: return 1
    elif x<2000: return 2
    elif x<3000: return 3
    elif x<4000: return 4
    elif x<5000: return 5
    elif x<6000: return 6
    elif x<7000: return 7
    elif x<8000: return 8
    elif x<9000: return 9
    elif x<10000: return 10
    return 11
faixas=dados.apply(faixasph,axis=1)
dados['faixas']=faixas
dados.faixas=dados.faixas.astype('category')
dados.faixas.cat.categories=(['1000','2000','3000','4000','5000','6000','7000','8000','9000','10000','+10000'])
```

```python
pd.crosstab(dados.faixas,dados.V06111).plot(kind='bar',stacked = False)
plt.xlabel("Renda")
plt.ylabel("População acessando a Internet")
```
