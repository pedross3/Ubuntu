```python
#Vamos agora criar uma lista de tempo com um delta t pequeno. Vamos usar uma lista de 0 a 2s com delta t de 0.001s. 
#Lembra-se do módulo "arange" do "numpy"?
tempo = np.arange(0,1.147,0.001)
```

```python
#Vamos agora executar o ODEINT para a obtenção da solução numérica do nosso sistema de equações diferenciais
from scipy.integrate import odeint
from math import pi
from math import sin
from math import cos
import matplotlib.pyplot as plt
%matplotlib inline

#Execute o ODEINT e plote o gráfico de Y(t) por X(t)
a = 35*pi/180
g = 10
x0 = 0
y0 = 0
vx0 = 10*cos(a)
vy0 = 10*sin(a)
Y0 = [x0,vx0,y0,vy0]
Sol = odeint(Eqdif, Y0, tempo)


plt.plot(tempo, Sol[:,0])
plt.xlabel("VelocidadeEm X")
plt.grid()
plt.show()
plt.plot(tempo, Sol[:,1])
plt.xlabel("Aceleração Em X")
plt.grid()
plt.show()
plt.plot(tempo, Sol[:,2])
plt.xlabel("Velocidade em Y")
plt.grid()
plt.show()
plt.plot(tempo, Sol[:,3])
plt.xlabel("Aceleração Em Y")
plt.grid()
plt.show()
```
